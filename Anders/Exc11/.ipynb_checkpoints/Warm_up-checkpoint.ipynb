# Chebyshevs polynomials

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def TnFactory(n):
    """
    Calculates the nth Chebyshev polynomial using the recursive relation:
        T_n = 2x * T_{n - 1} - T_{n - 2}
    :return: the nth Chebyshev polynomial
    :rtype: Polynomial
    """
    if n == 0:
        def T_n(x):
            return 1
    elif n == 1:
        def T_n(x):
            return x
    else:
        def T_n(x):
            return 2*x*TnFacctory(n-1)(x)-TnFacctory(n-2)(x)
    return T_n
```

```python
# Plot
x_start, x_end = -1, 1 # x-axis range
pols = 5               # Number of polynomials to plot

x = np.linspace(x_start,x_end,200)       
plt.figure(figsize=(5,4),dpi=400)
for i in range(1,pols+1):
    plt.plot(x, TnFactory(i)(x), label='$T_{}(x)$'.format(i))
plt.xlabel('$x$')
plt.ylabel('$T_n(x)$')
plt.title('First 5 Chebyshev polynomials, $T_n(x)$', fontsize=16)
plt.xlim([x_start,x_end])
plt.ylim([x_start,x_end])
plt.legend(loc='lower right', frameon=True)
```
