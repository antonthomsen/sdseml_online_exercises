**Problem description:**
Chebyshevs polynomials of the ﬁrst kind are deﬁned recursively by

$T_0(x) = 1$

$T_1(x) = x$

$T_n(x) = 2xT_{n−1}(x)−T_{n−2}(x)$

Write a factory function $TnFactory(n)$ that returns as its output the function $Tn(x)$.

Hint: you need to reactivate your knowledge on recursive functions!

Bonus: plot on $x ∈ (−1,1)$ the ﬁrst ﬁve Chebyshev polynomials $T1,...,T5$.

*Attempt:*

```python
# Defining a function
def TnFactory(n):
    # T_0
    if n == 0:
        return lambda x: 1
    # T_1
    if n == 1:
        return lambda x: x
    # T_n, returning for n-1 and n-2 with x value
    else:
        return lambda x: 2*x*TnFactory(n-1)(x)-TnFactory(n-2)(x)


import matplotlib.pyplot as plt
print('Tuple length', len(TnFactory(1)(x)))

for i in range(1,6):
        plt.plot(x, TnFactory(i)(x))

```
