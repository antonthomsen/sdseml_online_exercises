1. Exogenous/Contextual effects: Group characteristics effect on the individual. Group characteristics are often defined as the average value within the group.
2. Endogenous effects
3. Correlated effect: Like-minded people naturally group together and affect each other.  

Difficult to identify because of underidentification. We observe a single outcome that can be cause by the three social effect simultanously. Hence, for identification we would need outcomes in which only one effect is 'turned on' at the time (the other factors would have to be controlled for).
