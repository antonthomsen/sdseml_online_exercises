```python
# Exercise 11 warmup
def TnFactory(x,n):
    if n==0:
        return 1
    if n==1:
         return x
    else:
         return 2*x*TnFactory(x,n-1) - TnFactory(x,n-2) 

def TnFactory_x(n):
    return lambda x: TnFactory(x, n)

# Plotting
for n in range(1,6):
    f = TnFactory_x(n)
    m = [f(x) for x in np.arange(-1,1,0.01)]
    plt.plot(np.arange(-1,1,0.01), m)
```
