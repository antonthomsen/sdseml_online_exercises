```python
import numpy as np
X = np.linspace(-1, 1, 100)
Tn1 = []
Tn2 = []
Tn3 = []
Tn4 = []
Tn5 = []
def TnFactory(n,x):
        if n==0:
            return 1
        if n==1:
            return x
        else:
            return 2*x*TnFactory(n-1,x)-TnFactory(n-2,x)
```

```python
TnFactory(0,1), TnFactory(1,1), TnFactory(3,1), TnFactory(4,1)
```

```python
TnFactory(0,0.5), TnFactory(1,0.5), TnFactory(3,0.5), TnFactory(4,0.5)
```

```python
TnFactory(0,0), TnFactory(1,0), TnFactory(3,0), TnFactory(4,0)
```

```python
TnFactory(0,-0.5), TnFactory(1,-0.5), TnFactory(3,-0.5), TnFactory(4,-0.5)
```

```python
TnFactory(0,-1), TnFactory(1,-1), TnFactory(3,-1), TnFactory(4,-1)
```

```python
for i in X:
    Tn1.append(TnFactory(1,i))
    Tn2.append(TnFactory(2,i))
    Tn3.append(TnFactory(3,i))
    Tn4.append(TnFactory(4,i))
    Tn5.append(TnFactory(5,i))
```

```python
import matplotlib.pyplot as plt

plt.plot(Tn1, X)
plt.plot(Tn2, X)
plt.plot(Tn3, X)
plt.plot(Tn4, X)
plt.plot(Tn5, X)
```
