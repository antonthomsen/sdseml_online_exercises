```python
#Q2: All a,b pairs that makes a Pythagorean triple

def p_tri(n):
    for b in range(n):
        for a in range(1, b):
            c =math.sqrt(a**2 + b**2)
            if c % 1 == 0:
                print(a, b)
            
p_tri(10000)
```
