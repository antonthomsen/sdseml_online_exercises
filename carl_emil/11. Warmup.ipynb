Chebyshevs polynomials of the first kind are dined recursively by

\begin{align}
T_0(x) &= 1 \\
T_1(x) &= x \\
T_n(x) &= 2 x T_{n-1}(x) - T_{n-2} (x)
\end{align}

Write a *factory function* `TnFactory(n)` that returns as its output the *function* T_n(x). 

*I implemented the code below with help from: https://people.sc.fsu.edu/~jburkardt/py_src/chebyshev_polynomial/chebyshev_polynomial.py*

```python
import numpy as np
import matplotlib.pyplot as plt
from sys import exit

def Tn_factory(m, n, x):
    if (n < 0):
        exit('Error')

    v = np.zeros([m, n + 1])
    
    for i in range(0, m):
        v[i, 0] = 1.0
    
    if (n == 0):
        return v
    
    for i in range(0, m):
        v[i, 1] = x[i]
        
        for i in range(0, m):
            for j in range (2, n + 1):
                v[i, j] = 2.0 * x[i] * v[i, j-1] - v[i, j-2]
    
    return v

chebyshevs = Tn_factory(200, 5, np.linspace(-1, 1, 200))
```

```python
fig, ax = plt.subplots(figsize=(14,8))

with plt.style.context('fivethirtyeight'):
    for n in range(0, 6):
        ax.plot(np.linspace(-1, 1, 200), chebyshevs[:, n], label = f'$T_{n} (x)$')
    
plt.legend(loc = 7, prop={'size': 20}, bbox_to_anchor=(1.2, 0.5))
plt.tight_layout()
plt.ylabel('$T_n(x)$', size = 22)
plt.xlabel('$x$', size = 22)
plt.xticks(size = 12)
plt.yticks(size = 12)
plt.title('Chebyshevs polynomials of the first kind $(n=5)$', size = 26)
plt.show()
```
