```python
import matplotlib.pylab as plt
import numpy as np

def TnFactory(n,x):
    if n == 0:
        return 1
    if n == 1:
        return x
    return 2*TnFactory(n-1,x) - TnFactory(n-2,x)

X = np.linspace(-1,1,20)

for m in [1,2,3,4,5]:
    Y=[]
    for z in X:
        Y.append(TnFactory(m,z))
    plt.plot(X,Y)
```
